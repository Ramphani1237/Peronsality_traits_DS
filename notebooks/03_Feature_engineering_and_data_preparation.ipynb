# 3. Feature Engineering & Data Preparation

```python
import pandas as pd
df = pd.read_csv('../assets/raw/essaytrain.csv', encoding='latin1')

traits = ['cOPN', 'cCON', 'cEXT', 'cAGR', 'cNEU']
df_numeric = df.copy()
df_numeric[traits] = df_numeric[traits].replace({'y': 1, 'n': 0})
df_numeric['word_count'] = df_numeric['TEXT'].str.split().str.len()
df_numeric['char_count'] = df_numeric['TEXT'].str.len()
df_numeric['total_trait_score'] = df_numeric[traits].sum(axis=1)
print(df_numeric[traits].describe())
print(df_numeric[['word_count', 'char_count']].describe())
X = df_numeric[['TEXT', 'word_count', 'char_count']]
y = df_numeric[traits]
```
