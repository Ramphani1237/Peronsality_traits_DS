```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load dataset
df = pd.read_csv('../data/raw/essaytrain.csv', encoding='latin1')

# Prepare traits
# =========================
traits = ['cEXT', 'cNEU', 'cAGR', 'cCON', 'cOPN']

# Convert y/n to 1/0 for numeric analysis
df_numeric = df[traits].replace({'y':1, 'n':0}).astype(int)

# Create folder for plots
plots_dir = '../data/plots'
os.makedirs(plots_dir, exist_ok=True)

# =========================
# 1️⃣ Trait-level Bar and Pie Charts
# =========================
for trait in traits:
    counts = df[trait].value_counts()
    total = counts.sum()
    
    # Bar chart (counts)
    plt.figure(figsize=(6,4))
    sns.barplot(x=counts.index, y=counts.values)
    plt.title(f'{trait} - Counts of Yes/No')
    plt.ylabel('Number of Participants')
    plt.xlabel('Response')
    plt.tight_layout()
    plt.savefig(os.path.join(plots_dir, f'{trait}_bar_counts.png'))
    plt.close()
    
    # Pie chart (percentages)
    plt.figure(figsize=(5,5))
    plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=['#4CAF50','#FF5722'])
    plt.title(f'{trait} - Percentage of Responses')
    plt.savefig(os.path.join(plots_dir, f'{trait}_pie_percent.png'))
    plt.close()
    
    print(f"Value counts for {trait}:")
    print(counts)
    print("\n")

# =========================
# 2️⃣ Correlation Heatmap Between Traits
# =========================
plt.figure(figsize=(8,6))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between OCEAN Traits')
plt.tight_layout()
plt.savefig(os.path.join(plots_dir, 'traits_correlation_heatmap.png'))
plt.close()

print("All plots saved in:", plots_dir)
```
